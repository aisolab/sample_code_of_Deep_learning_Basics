### Class  생성

```python
## Class 생성
class Man: # Man이라는 이름의 class를 생성하려면
    def __init__(self, name): # 먼저 __init__ 메서드 (생성자)를 활용한다. self는 해당 Class에서 나오는 instance를 명시적으로 한 번 선언한 것
        self.name = name      # instance variable을 선언한다. instance variable은 instance 별로 저장되는 variable
        print('Initialized!') # 결국 이 부분은 name 인자를 받아서 Man Class의 instance variable인 self.name 값으로 인자로 받은 name을 할당
    
    def hello(self): # Man Class의 hello 메서드를 정의
        print('Hello ' + self.name + '!')
    
    def goodbye(self): # Man Class의 goodbye 메서드를 정의
        print('Good-bye ' + self.name + '!')
```

```python
m = Man('David silver') # David silver AlphaGo의 제작자 중 한 명입니다.
m.hello()
m.goodbye()
```

### numpy 활용 

```python
import numpy as np
# 1차원 배열생성(1차원 배열 : 벡터, 2차원 배열 : 행렬, 3차원 이상의 배열 : 텐서 또는 다차원배열)
a = np.array([1, 2, 3])
print(a) # a 출력
type(a) # a의 class 확인 
a.shape # a의 모양확인
```

```python
# 1차원 배열의 산술연산 (element-wise)
b = np.array([2, 4, 6])
a + b # 나누기
a - b # 빼기
a * b # 곱하기
a ** b # 승
a / b # 몫
a % b # 나머지
b / 2 # numpy 배열과 수치 하나의 연산 broadcasting 적용
```

```python
# 2차원 배열 생성
c = np.array([[1, 2], [3, 4]])
c.shape
c.dtype
```

```python
# 2차원 배열의 산술연산 (element-wise)
d = np.array([[3, 0], [0, 6]])
c + d
c - d
c * d
c ** d
c / d
c % d
d + 10
```

```python
# broadcasting
a = np.array([1,2,3])
b = 5
a + b
c = np.array([[1, 2], [3, 4]])
d = np.array([10, 20])
c * d # 1차원 배열인 d가 2차원 배열 c와 같은 형상으로 변형된 후 원소별 연산이 이루어짐
```

```python
# 원소접근
x = np.array([[51, 55], [14, 19], [0, 4]])
print(x)
x[0]
x[0, 0]
x[0, 1]
x[x >= 50]
```

```python
x = x.flatten()
print(x)
x[x >= 50]
```

```python
## matplotlib 활용
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# first example
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# second example
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle = '--', label = 'cos') # cos 함수는 점선으로 그리기
plt.xlabel('x') # x축 이름
plt.xlabel('y') # y축 이름
plt.title('sin & cos') # 그림제목
plt.legend()
plt.show()
```

```python
# third example : 이미지표시하기
from skimage import io
img = io.imread('http://i.imgur.com/MOv7KaS.jpg')
# third example : 이미지 표시하기
fig = plt.figure()
fig.add_subplot(2,2,1)
plt.imshow(img) # original
fig.add_subplot(2,2,2)
plt.imshow(img[:,:,0]) # red channel
fig.add_subplot(2,2,3)
plt.imshow(img[:,:,1]) # green channel
fig.add_subplot(2,2,4)
plt.imshow(img[:,:,2]) # blue channel
plt.show()
```
